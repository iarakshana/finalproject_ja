```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile 
pd.set_option('display.max_rows', None)

df1 = pd.read_excel('2014Senate.xlsx')

df_vote = df1[["GENERAL %" , "CANDIDATE NAME", "STATE"]]

#We realistically only needed to consider candidates who got more than around 17% of the vote. Coincidentally this cutoff worked for both 2014 and 2012. 
df_real_vote = df_vote[(df1["GENERAL %"]) > .173]

#A lot of states use 'Scattered' to describe the variety of candidates who got very few votes so removing this was helpful. 
df_2014_Senate = df_real_vote[df_real_vote["CANDIDATE NAME"].str.contains("Scattered")==False]

#We reset the index so we could merge this with the Google Trends dataframe. 
df_2014S = df_2014_Senate.reset_index(inplace = False)


#2014 was the only year that had separate datasets for the House and the Senate; for the other two we had to split up the one dataset into two usable ones. 
df2 = pd.read_excel('2014House.xlsx')

df_vote2 = df2[["GENERAL %" , "D" , "CANDIDATE NAME" , "STATE"]]

#Sadly no reasonable cutoff could be defined for the House 
df_real_vote2 = df_vote2[(df2["GENERAL %"]) >= 0]

df_2014_House = df_real_vote2[df_real_vote2["CANDIDATE NAME"].str.contains("Scattered")==False] 

df_2014H = df_2014_House.reset_index(inplace = False)


df3 = pd.read_excel('2012.xlsx')

df_vote3 = df3[["GENERAL %" , "CANDIDATE NAME" , "D" , "STATE"]]

#For 2012 and 2010 we had to separate the Senate races from the House by the cutoff and by the fact that each senate race was displayed with 'S' as the district ('D). 
df_real_vote3S = df_vote3[(df3["GENERAL %"]) > .17]

df_real_vote3H = df_vote3[(df3["GENERAL %"]) >= 0]

df_Sen = df_real_vote3S[df_real_vote3S["D"].str.contains("S")]

df_House = df_real_vote3H[df_real_vote3H["D"].str.contains("S")==False]

df_2012_Senate = df_Sen[df_Sen["CANDIDATE NAME"].str.contains("Scattered")==False]

df_2012_House = df_House[df_House["CANDIDATE NAME"].str.contains("Scattered")==False]

df_2012S = df_2012_Senate.reset_index(inplace = False)

df_2012H = df_2012_House.reset_index(inplace = False)



df4 = pd.read_excel('2010.xlsx')

df_vote4 = df4[["GENERAL %" , "CANDIDATE NAME (Last, First)" , "DISTRICT" , "STATE"]]

#Once we started using the for loop for pytrends we could no longer rely on cutoffs to make these datasets match up. We needed two names for every race to match the for loop so if a candidate had no real opponent, we added in 'All Others' with 0% of the vote. 
df_real_vote4S = df_vote4[(df4["GENERAL %"]) >= 0]

df_real_vote4H = df_vote4[(df4["GENERAL %"]) >= 0]

df_Senate = df_real_vote4S[df_real_vote4S["DISTRICT"].str.contains("S", na=False)]

df_house = df_real_vote4H[df_real_vote4H["DISTRICT"].str.contains("S", na=False)==False]

df_2010_Senate = df_Senate[df_Senate["CANDIDATE NAME (Last, First)"].str.contains("Scattered")==False]

df_2010_House = df_house[df_house["CANDIDATE NAME (Last, First)"].str.contains("Scattered")==False]

df_2010S = df_2010_Senate.reset_index(inplace = False)

df_2010H = df_2010_House.reset_index(inplace = False)

```
