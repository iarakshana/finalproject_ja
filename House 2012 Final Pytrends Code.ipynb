```python
#For Names 1
#We created a function that would run throught the array of names in our list to get the interest over time data. We then obtained the
#proportion of vote for each candidate and appended it to our list
gtrends2012H1 = []

def searchState(kw_list):
    pytrend.build_payload(kw_list, geo = 'US',  timeframe = '2012-01-01 2012-11-06')
    iot_2012H1 = pytrend.interest_over_time()
    person1 = (sum(iot_2012H1[kw_list[0]]))/ ((sum(iot_2012H1[kw_list[0]])) + (sum(iot_2012H1[kw_list[1]])))
    person2 = 1-person1
    x = (person1, person2)
    gtrends2012H1.append(x)
  #prints each
  #print(x)


kw_list = [searchState(n) for n in names1] # List Comprehension so that the list can go through each set of names of the total names1 list
```
