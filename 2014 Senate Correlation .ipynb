```python
df1 = pd.read_excel('2014Senate.xlsx')

df_vote = df1[["GENERAL %" , "CANDIDATE NAME", "STATE"]]

df_real_vote = df_vote[(df1["GENERAL %"]) > .173]

df_2014_Senate = df_real_vote[df_real_vote["CANDIDATE NAME"].str.contains("Scattered")==False]

df_2014S = df_2014_Senate.reset_index(inplace = False)
```

```python
df_gtrends_2014S = pd.DataFrame(np.array(gtrends2014S))
df_gtrends_2014S = df_gtrends_2014S.rename(columns = {0 : 'Percent Searched'})

df_2014S = df_2014S[["CANDIDATE NAME" , "GENERAL %"]]

df_combined = df_2014S.join(df_gtrends_2014S, how = "left")
```

```python
%matplotlib inline

#This graph demonstrates some of the difficulty with using the pytrends data; candidates such as Kevin Wade who shared a name with a popular figure (screenwriter in this case) didn't show up which meant that their opponent got 100% of the searches. When this didn't occur the proportion of searches lined up with the proportion of votes quite nicely. 
#Cases such as Kevin Wade's likely contributed to several of the outliers in the following graph. 
df_combined.head(9).plot(kind = 'bar', x = "CANDIDATE NAME")
```

```python
import seaborn as sns 

sns.regplot( x = "Percent Searched", y = "GENERAL %" , data = df_combined)
```

```python
from scipy import stats
import statsmodels.api as sm
import statsmodels.formula.api as smf 
import statsmodels.regression.mixed_linear_model as sm_mlm

ols = smf.ols(formula = "GP ~ PS", data = df_combined.rename(columns = {"GENERAL %" : "GP", "Percent Searched" : "PS"}))
model = ols.fit()
model.summary()
```

```python
sns.regplot(x = df_combined["Percent Searched"], y = pd.Series(model.resid, name = "Residual"))
#There are several cases where the residuals are relatively high (probably when pytrends couldn't find the losing candidate) and also many where the residuals are relatively low, implying that there were some races where the loser was much more popular than the winner. 
```
