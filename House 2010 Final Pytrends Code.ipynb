```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np

pytrend = TrendReq()
```

```python
#For Names 1
gtrends2010H1 = []

def searchState(kw_list):
    pytrend.build_payload(kw_list, geo = 'US',  timeframe = '2010-01-01 2010-11-02')
    iot_2010H1 = pytrend.interest_over_time()
    person1 = (sum(iot_2010H1[kw_list[0]]))/ ((sum(iot_2010H1[kw_list[0]])) + (sum(iot_2010H1[kw_list[1]])))
    person2 = 1-person1
    x = (person1, person2)
    gtrends2010H1.append(x)
  #prints each
  #print(x)


kw_list = [searchState(n) for n in names1]
```

```python
from pprint import pprint
gtrends10H1 = ([item for c_list in gtrends2010H1 for item in c_list])
# https://stackoverflow.com/questions/22882143/split-lists-within-list
gtrends10H1 = pd.DataFrame(np.array(gtrends10H1))
gtrends10H1.rename(columns={0 : 'Percent'})
gtrends10H1
```
