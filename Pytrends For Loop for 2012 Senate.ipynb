```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np

pytrend = TrendReq()

names = [['Jeff Flake','Richard Carmona'], ['Dianne Feinstein','Elizabeth Emken'], ['Chris Murphy', 'Linda E McMahon'], ['Tom Carper', 'Kevin L. Wade'], ['Bill Nelson', 'Connie Mack'], ['Mazie Hirono', 'Linda Lingle'], ['Joe Donnelly', 'Richard E Mourdock'], ['Angus King', 'Charles E Summers'], ['Ben Cardin', 'Daniel John Bongino'], ['Elizabeth Warren', 'Scott P Brown'], ['Debbie Stabenow', 'Pete Hoekstra'], ['Amy Klobuchar', 'Kurt Bills'], ['Roger F Wicker', 'Albert N Gore'], ['Claire McCaskill', 'Todd Akin'], ['Jon Tester', 'Denny Rehberg'], ['Deb Fischer', 'Bob Kerrey'], ['Dean Heller', 'Shelley Berkley'], ['Robert Menendez', 'Joe Kyrillos'], ['Martin Heinrich', 'Heather A Wilson'], ['Kirsten Gillibrand', 'Wendy Long'], ['Heidi Heitkamp', 'Rick Berg'], ['Sherrod Brown', 'Josh Mandel'], ['Bob Casey', 'Tom Smith'], ['Sheldon Whitehouse', 'Benjamin Barrett Hinckley'], ['Bob Corker', 'Mark E Clayton'], ['Ted Cruz', 'Paul Sadler'], ['Orrin Hatch', 'Scott Howell'], ['Bernie Sanders', 'John MacGovern'], ['Timothy Kaine', 'George F Allen'], ['Maria Cantwell', 'Michael Baumgartner'], ['Joe Manchin', 'John R Raese'], ['Tammy Baldwin', 'Tommy G Thompson'], ['John Barrasso', 'Tim Chestnut']]


gtrends2012S = []

def searchState(kw_list):
  pytrend.build_payload(kw_list, geo = 'US',  timeframe = '2012-01-01 2012-11-06')
  iot_2012 = pytrend.interest_over_time()
  person1 = (sum(iot_2012[kw_list[0]]))/ ((sum(iot_2012[kw_list[0]])) + (sum(iot_2012[kw_list[1]])))
  person2 = 1-person1
  x = (person1, person2)
  gtrends2012S.append(x)
  #prints each
  #print(x)


kw_list = [searchState(n) for n in names]

#prints all
#print(gtrends2012S)
type(gtrends2012S)
```
