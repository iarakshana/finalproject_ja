```python
from pytrends.request import TrendReq
import pandas as pd
import numpy as np

# Login to Google. Only need to run this once, the rest of requests will use the same session.
pytrend = TrendReq()

#Alabama
#2014
#We start with Alabama and go on alphabetically in order so that when appended the candidates are in line with
#the indices in the FEC database

pytrend.build_payload(kw_list=["Jefferson Sessions"], geo = 'US',  timeframe = '2014-01-01 2014-11-04')
iot_2014_al = pytrend.interest_over_time()
JSessp = (sum(iot_2014_al["Jefferson Sessions"]))/ ((sum(iot_2014_al["Jefferson Sessions"])))
gtrends2014S = pd.Series([JSessp])

print(gtrends2014S)
```
