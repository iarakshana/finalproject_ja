```python
#For Names 1
gtrends2012H1 = []

def searchState(kw_list):
    pytrend.build_payload(kw_list, geo = 'US',  timeframe = '2012-01-01 2012-11-06')
    iot_2012H1 = pytrend.interest_over_time()
    person1 = (sum(iot_2012H1[kw_list[0]]))/ ((sum(iot_2012H1[kw_list[0]])) + (sum(iot_2012H1[kw_list[1]])))
    person2 = 1-person1
    x = (person1, person2)
    gtrends2012H1.append(x)
  #prints each
  #print(x)


kw_list = [searchState(n) for n in names1]
```
